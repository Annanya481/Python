```
#Program to find PI upto N digits
import math

limit = int(input('Enter limit of PI (less than 25)'))

if limit>25:
  print('Too large')
else:
  print('%.*f' % (limit, math.pi))
```

```
#Program to find e upto N digits
import math

limit = int(input('Enter limit of e(upto 25'))

if limit >25:
  print('Too large')
else:
  print('%.*f' % (limit, math.e))
```

```
#Program to generate Fibonacci sequence upto N digits
limit = int(input('Enter limit:  '))
a = 0
print(a)
b = 1
print(b)
ct = 3
while ct<=limit:
  c = a + b
  print(c)
  a = b
  b = c
  ct+=1
```

```
#Program to find prime factors of input number

def check_prime(n):
  for i in range(2, int(n/2)):
    if n % i == 0:
      return False
  return True

number = int(input('Enter a number'))
for i in range(2, number-1):
  if number%i == 0 and check_prime(i) == True:
    print(i)
```

```
#Program to generate prime numbers till the limit entered by the user
def check_prime(n):
  for i in range(2, int(n/2 + 1)):
    if n%i == 0:
      return False
  return True

num = int(input('Enter a limit'))
for i in range(2, num):
  if check_prime(i):
    print(i)
```

```
#Program to claculate the cost of tiling a floor when the cost and dimensions are entered by the user
width = float(input('Enter width of floor:  '))
height = float(input('Enter height of floor:  '))
cost = float(input('Enter cost:  '))
area = width * height
total = cost * area
print('Total cost = ',total)
```

```
#Program to convert a binary number to decimal 
decimal = 0
binaryNum = list(input('Enter a binary number:  '))
length = len(binaryNum)
length -= 1
while length>=0:
  decimal += (int(binaryNum[length]) * 2**length)
  length -= 1

print(decimal)
```

```
#Program to convert a decimal number to binary
def DecimaltoBinary(num):
  if num>1:
    DecimaltoBinary(num//2)
  print(num%2, end='')

decimal = int(input('Enter a number:  '))
DecimaltoBinary(decimal)
```

```
#Calculator
def Addition(num1, num2):
  return num1 + num2
def Subtraction(num1, num2):
  if num1>num2:
    return num1-num2
  return num2-num1
def Multiplication(num1, num2):
  return num1*num2
def Division(num1, num2):
  return num1//num2
def Remainder(num1, num2):
  return num1%num2

a = int(input('Enter number 1:  '))
b = int(input('Enter number 2:  '))
op = input('Enter operation:  ')
if op=='+':
  print(Addition(a, b))
elif op=='-':
  print(Subtraction(a, b))
elif op=='*':
  print(Multiplication(a, b))
elif op=='/':
  print(Division(a, b))
elif op=='%':
  print(Remainder(a, b))
else:
  print('Wrong operation')
```
